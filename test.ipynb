```python
import json
import pandas as pd

def json_to_csv(json_file_path, csv_file_path):
    """
    Converts a JSON file containing a list of chemical compound data into a CSV file.

    Args:
        json_file_path (str): Path to the input JSON file.
        csv_file_path (str): Path to the output CSV file.
    """

    with open(json_file_path, 'r') as f:
        data = json.load(f)

    # Handle potential variations in JSON structure
    if isinstance(data, list):  # List of dictionaries
        df = pd.DataFrame(data)
    elif isinstance(data, dict):  # Single dictionary
        df = pd.DataFrame([data]) 
    else:
        raise ValueError("Unsupported JSON format. Expected a list or a single dictionary.")

    # Optional: Select specific columns
    # columns_to_keep = ['Name', 'SMILES', 'InChI']
    # df = df[columns_to_keep]

    # Save to CSV
    df.to_csv(csv_file_path, index=False) 

# Usage Example
json_file_path = '/home/mescalin/yitao/Documents/Code/CRN_IMA/namesmilesinchi_pretty.json'
csv_file_path = '/home/mescalin/yitao/Documents/Code/CRN_IMA/namesmilesinchi_pretty.csv'
json_to_csv(json_file_path, csv_file_path)
```

```python
import pandas as pd
import rdkit.Chem as Chem

def is_disconnected(smiDFS):
    mol = Chem.MolFromSmiles(smiDFS)
    fragments = Chem.GetMolFrags(mol, asMols=True)
    return len(fragments) > 1

def delete_disconnected_mols(csv_file_path):
    df = pd.read_csv(csv_file_path)
    df.columns = df.columns.str.lower()
    # Filter out rows with disconnected molecules
    df_filtered = df[~df['smiles'].apply(is_disconnected)]

    # Overwrite the original CSV file
    df_filtered.to_csv(csv_file_path, index=False)

# Example Usage
csv_file_path = "/home/mescalin/yitao/Documents/Code/CRN_IMA/namesmilesinchi_pretty.csv"  # Replace with your CSV file path
delete_disconnected_mols(csv_file_path)
```

```python


def extract_bondtype(rxns_arr, num_sample):
    reactant_bonds = set()
    products_bonds = set()
    rxn_samples = np.random.choice(rxns_arr, num_sample, replace=False)

    for r in rxn_samples:
        rxns = AllChem.ReactionFromSmarts(r)
        for reactant in rxns.GetReactants():
            for bondr in reactant.GetBonds():
                bondr_label = bondr.GetBondTypeAsDouble()
                bondr_type = bondr.GetBondType()
                reactant_bonds.add((bondr_label,bondr_type))
        for product in rxns.GetProducts():
            for bondp in product.GetBonds():
                bondp_label = bondp.GetBondTypeAsDouble()
                bondp_type = bondp.GetBondType()
                products_bonds.add((bondp_label,bondp_type))

    bonds = reactant_bonds | products_bonds

    bonds_array = np.array(list(bonds), dtype=[('value1', float), ('value2', 'U20')])


    unique_bonds_array = np.unique(bonds_array, return_index=False)
 

    return unique_bonds_array


unique_bonds = extract_bondtype(rxn_smt_array[:,0], 349000)  

unique_bonds
```

```python
from rdkit.Chem import AllChem

def smarts_to_gml(reaction_smarts_array):
    reaction_smarts, reaction_name = reaction_smarts_array[0],reaction_smarts_array[1]

    reaction = AllChem.ReactionFromSmarts(reaction_smarts)


    gml_rule = {
        "ruleID": f'"{reaction_name}"',
        "left": {"edges": []},
        "context": {"nodes": [], "edges": []},
        "right": {"edges": []}
    }

    atom_id_map = {}
    atom_id_counter = 1
    bonds = {1: "-", 2: "=", 3: "#", 1.5: ":"}
    for reactant in reaction.GetReactants():
        for atom in reactant.GetAtoms():
            atom_id = atom.GetAtomMapNum()
            if atom_id is not None:
                atom_id_map[atom_id] = atom.GetSymbol()
                atom_id_counter += 1


    # reactant_bonds = set()
    # product_bonds = set()

    # for reactant in reaction.GetReactants():
    #     for bond in reactant.GetBonds():
    #         begin_atom = bond.GetBeginAtom().GetAtomMapNum()
    #         end_atom = bond.GetEndAtom().GetAtomMapNum()
    #         if begin_atom is not None and end_atom is not None:
    #             bond_type = bond.GetBondTypeAsDouble()
    #             reactant_bonds.add((begin_atom, end_atom, bond_type))

    # for product in reaction.GetProducts():
    #     for bond in product.GetBonds():
    #         begin_atom = bond.GetBeginAtom().GetAtomMapNum()
    #         end_atom = bond.GetEndAtom().GetAtomMapNum()
    #         if begin_atom is not None and end_atom is not None:
    #             bond_type = bond.GetBondTypeAsDouble()
    #             product_bonds.add((begin_atom, end_atom, bond_type))

    reactant_bonds = {(min(b.GetBeginAtom().GetAtomMapNum(), b.GetEndAtom().GetAtomMapNum()), 
                       max(b.GetBeginAtom().GetAtomMapNum(), b.GetEndAtom().GetAtomMapNum()), b.GetBondTypeAsDouble()) 
                       for reactant in reaction.GetReactants() 
                       for b in reactant.GetBonds() 
                       if b.GetBeginAtom().GetAtomMapNum() is not None and b.GetEndAtom().GetAtomMapNum() is not None}
    product_bonds = {(min(b.GetBeginAtom().GetAtomMapNum(), b.GetEndAtom().GetAtomMapNum()), 
                      max(b.GetBeginAtom().GetAtomMapNum(), b.GetEndAtom().GetAtomMapNum()), b.GetBondTypeAsDouble()) 
                      for product in reaction.GetProducts() 
                      for b in product.GetBonds() 
                      if b.GetBeginAtom().GetAtomMapNum() is not None and b.GetEndAtom().GetAtomMapNum() is not None}
    
    print(f"Reactant bonds: {reactant_bonds}")

    print(f"Products bonds: {product_bonds}")

    unchanged_bonds = reactant_bonds.intersection(product_bonds)

    print(f"Unchanged bonds: {unchanged_bonds}")
    print(f"atom id map before sort: {atom_id_map}")

    atom_id_map = dict(sorted(atom_id_map.items(), key=lambda item: item[0]))

    print(f"Atom ID map: {atom_id_map}")
    for atom_id, label in atom_id_map.items():
        gml_rule["context"]["nodes"].append({
            "id": atom_id,
            "label": label
        })

    for bond in unchanged_bonds:
        source, target, label = bond
        bond_label = bonds[label]
        gml_rule["context"]["edges"].append({
            "source": source,
            "target": target,
            "label": bond_label
        })


    changed_reactant_bonds = sorted(reactant_bonds - unchanged_bonds, key=lambda x: x[0])

    print(f"Changed reactant bonds:{changed_reactant_bonds}")
    changed_product_bonds = sorted(product_bonds - unchanged_bonds, key=lambda x: x[0])
    print(f"Changed product bonds: {changed_product_bonds}")
    for source, target, label in changed_reactant_bonds:
        bond_label = bonds[label]
        gml_rule["left"]["edges"].append({
            "source": source,
            "target": target,
            "label": bond_label
        })

    for source, target, label in changed_product_bonds:
        bond_label =  bonds[label]
        gml_rule["right"]["edges"].append({
            "source": source,
            "target": target,
            "label": bond_label
        })

    gml_format = f"rule[\n ruleID {gml_rule['ruleID']}"
    for key in ["left", "context", "right"]:
        gml_format += f" \n {key} ["
        if "nodes" in gml_rule[key]:
            for item in gml_rule[key]["nodes"]:
                gml_format += f"""\n   node [ id {item['id']} label "{item['label']}" ]"""
        for item in sorted(gml_rule[key]["edges"], key=lambda x: (x['source'],x['target'])):
            gml_format += f"""\n   edge [ source {item['source']} target {item['target']} label "{item['label']}"]"""
        gml_format += " ]"
    gml_format += " ]"


    # gml_format = f"rule[\n ruleID {gml_rule['ruleID']}]\n"
    # for key in ["left", "context", "right"]:
    #     gml_format += f"{key} [\n"
    #     if "nodes" in gml_rule[key]:
    #         gml_format += "\n".join([f"  node [ id {item['id']} label '{item['label']}']" for item in gml_rule[key]["nodes"]])
    #     gml_format += "\n".join([f" edge [ source {item['source']} target {item['target']} label '{item['label']}']" for item in gml_rule[key]["edges"]])
    #     gml_format += "]\n"

    return gml_format

# Example reaction SMARTS and reaction name
reaction_smarts = "[H:1][O:20][C:3][C:100][H:5]>>[H:1][O:20][H:5].[C:3]=[C:100]"
reaction_name = "Hydrogenation"
rxn_arr = [reaction_smarts,reaction_name]
# Convert the reaction SMARTS into GML format rule
gml_format_rule = smarts_to_gml(rxn_arr)

# Print the resulting GML format rule
print(gml_format_rule)




# RXN.ReactionFromSmarts(reaction_smarts, useSmiles=True)
```
